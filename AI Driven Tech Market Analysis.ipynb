```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Technology_and_AI_Companies_Market_Dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#AI Revenue Dependency Distribution ==> HISTOGRAM
plt.figure(figsize=(8,5))
plt.hist(df["AI_Revenue_Percent"], bins=10)
plt.title("Distribution of AI Revenue Percentage")
plt.xlabel("AI Revenue %")
plt.ylabel("Number of Companies")
plt.show()
```

```python
#Growth Rate by AI Focus Area ==> HORIZONTAL BAR
growth_focus = df.groupby("AI_Focus_Area")["Growth_Rate_2024_Percent"].mean().sort_values()
plt.figure(figsize=(10,5))
growth_focus.plot(kind="barh")
plt.title("Average Growth Rate by AI Focus Area")
plt.xlabel("Growth Rate (%)")
plt.show()
```

```python
#R&D Spending vs AI Patents ==> SCATTER
plt.figure(figsize=(8,5))
plt.scatter(
    df["R&D_Spending_Billion_USD"],
    df["AI_Patents_Count"]
)
plt.xlabel("R&D Spending (Billion USD)")
plt.ylabel("AI Patents Count")
plt.title("R&D Spending vs AI Innovation Output")
plt.show()
```

```python
#Company Size Segmentation
df["Company_Size"] = pd.cut(
    df["Market_Cap_Trillion_USD"],
    bins=[0, 1, 3, 6, 10],
    labels=["Small", "Mid", "Large", "Mega"]
)

size_dist = df["Company_Size"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(size_dist, labels=size_dist.index, autopct="%1.1f%%")
plt.title("Company Size Distribution by Market Cap")
plt.show()
```

```python
#Cloud Revenue vs AI Revenue ==> COMPARATIVE BAR
cloud_ai = df.groupby("Sector")[["Cloud_Revenue_Billion_USD", "AI_Revenue_Percent"]].mean()
cloud_ai.plot(kind="bar", figsize=(10,5))
plt.title("Cloud Revenue vs AI Revenue Dependency by Sector")
plt.show()
```

```python
#ESG Score Impact on AI Revenue
esg_ai = df.groupby("ESG_Score")[["AI_Revenue_Percent", "Cloud_Revenue_Billion_USD"]].sum()
esg_ai.plot(kind="area", figsize=(8,5))
plt.title("AI & Cloud Revenue by ESG Score")
plt.xlabel("ESG Score")
plt.ylabel("Revenue Contribution")
plt.show()
```

```python
print("Key Insights:")
print("- Market capitalization in the AI industry is heavily concentrated in Semiconductor and Software sectors, indicating their central role in AI infrastructure and applications.")
print("- AI-focused companies show varying levels of innovation efficiency, where higher R&D spending does not always translate into proportionally higher AI patent output.")
print("- AI revenue dependency is significantly higher in infrastructure- and chip-oriented AI focus areas compared to platform or application-based companies.")
print("- Companies with stronger ESG ratings tend to generate higher combined AI and cloud revenues, suggesting alignment between sustainable practices and long-term AI growth.")
print("- High growth rates do not consistently correlate with higher stock prices, highlighting the influence of market expectations, valuation cycles, and investor sentiment.")
print("- Revenue and market value are unevenly distributed, with a small number of companies contributing disproportionately to total AI market revenue.")
print("- Cloud revenue acts as a stabilizing revenue stream, particularly for large-cap technology firms with diversified AI portfolios.")
```
